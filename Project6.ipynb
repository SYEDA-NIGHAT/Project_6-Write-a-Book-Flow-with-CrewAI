```python
!pip install crewai google-generativeai python-dotenv litellm

import os
import google.generativeai as genai
from crewai import Agent, Task, Crew, Process
from dotenv import load_dotenv
import litellm

load_dotenv()

GOOGLE_API_KEY = "AIzaSyD1qYnrHUOIYYwYYw1Pr0_a4lCbAmAv6FM"
os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
litellm.api_key = GOOGLE_API_KEY

genai.configure(api_key=GOOGLE_API_KEY)


author = Agent(
    role="Creative Writer",
    goal="Write engaging and insightful book chapters.",
    backstory="A passionate writer with years of experience in fiction and non-fiction.",
    verbose=True,
    allow_delegation=True,
    model={"model": "gemini-1.5-pro", "litellm_provider": "google"}  # ✅ Correct model & provider
)

editor = Agent(
    role="Editor",
    goal="Refine and enhance the book's content for clarity and engagement.",
    backstory="A professional editor with a keen eye for detail and storytelling.",
    verbose=True,
    allow_delegation=True,
    model={"model": "gemini-1.5-pro", "litellm_provider": "google"}  # ✅ Same model & provider
)

write_chapters = Task(
    description="Write the initial drafts of the book chapters based on the given outline.",
    agent=author,
    expected_output="A detailed and well-structured draft of the book chapters."
)

edit_chapters = Task(
    description="Review and refine the book chapters for quality and coherence.",
    agent=editor,
    expected_output="A polished and refined version of the book chapters with improved clarity."
)

book_crew = Crew(
    agents=[author, editor],
    tasks=[write_chapters, edit_chapters],
    process=Process.sequential
)

try:
    result = book_crew.kickoff()
    print(result)
except Exception as e:
    print(f"Error: {e}")
```
